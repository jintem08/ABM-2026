```python
   import numpy as np

# 1) instellingen
W, H = 50, 50   # grid size
N = 1000        # burgers
B = 25          # textielbakken

# 2) random generator (reproduceerbaar)
rng = np.random.default_rng(42)

# 3) posities genereren
citizens_xy = np.column_stack([
    rng.integers(0, W, size=N),   # x-coördinaten
    rng.integers(0, H, size=N),   # y-coördinaten
])

bins_xy = np.column_stack([
    rng.integers(0, W, size=B),
    rng.integers(0, H, size=B),
])

print(citizens_xy.shape)  # (1000, 2)
print(bins_xy.shape)      # (25, 2)
```
